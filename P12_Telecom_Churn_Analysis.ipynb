#### 8 - Applying the Logistic Regression on the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,recall_score,precision_score,f1_score,roc_curve,roc_auc_score,classification_report
```

```python
def model_(model,X,y):
    tr_score = []
    ts_score = []
    for i in np.arange(0.1,1,0.1):
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=i,random_state=6)
        model.fit(X_train,y_train)
        tr_score.append(model.score(X_train,y_train))
        ts_score.append(model.score(X_test,y_test))
    plt.figure(figsize=(15,5))
    q = sns.lineplot(np.arange(0.1,1,0.1),tr_score,marker='*',label='Training Score')
    q = sns.lineplot(np.arange(0.1,1,0.1),ts_score,marker='+',label='Test Score')
```

```python
def model_score(model,X,y,test_size):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size,random_state=6)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print("Training Score of the Model is: ", model.score(X_train,y_train))
    print("Test Score of the Model is: ", model.score(X_test,y_test))
    print("Precision Score: ", precision_score(y_test,y_pred))
    print("Recall Score: ", recall_score(y_test,y_pred))
    print("F1 Score: ", f1_score(y_test,y_pred))
    print("")
    print("classification_report:")
    print(classification_report(y_test,y_pred))
```

```python
def model_only_score(model,X,y,test_size):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size,random_state=6)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print("Training Score of the Model is: ", model.score(X_train,y_train))
    print("Test Score of the Model is: ", model.score(X_test,y_test))
    print("Accuracy Score of the Model is: ", accuracy_score(y_test,y_pred))
```

```python
lr_model = LogisticRegression()
```

```python
model_(lr_model,X,y)
```

We can see from graph, test_size ideal here is 0.7

```python
model_score(lr_model,X,y,0.7)
```

#### 9 - Displaying the Confusion Matrix

```python
def get_con_mat(model,X,y,test_size):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size,random_state=6)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print(confusion_matrix(y_test,y_pred))
    plt.figure(figsize = (15,5))
    sns.heatmap(confusion_matrix(y_test,y_pred), annot = True);
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
get_con_mat(lr_model,X,y,0.7)
```

#### 12 - Tuning with the SVM model parameters i.e. different values of C

```python
tr_score = []
ts_score = []
for i in np.arange(5,100,5):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=6)
    svc=SVC(C=i)
    svc.fit(X_train,y_train)
    tr_score.append(svc.score(X_train,y_train))
    ts_score.append(svc.score(X_test,y_test))
plt.figure(figsize=(15,5))
q = sns.lineplot(np.arange(5,100,5),tr_score,marker='*',label='Training Score')
q = sns.lineplot(np.arange(5,100,5),ts_score,marker='+',label='Test Score')
```

If we take C = 65, it almost yields us best results

```python
svc=SVC(C=65)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Training Score: ",svc.score(X_train,y_train))
print("Testing Score: ",svc.score(X_test,y_test))
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```
